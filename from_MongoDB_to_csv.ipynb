## Данный ноутбук считывает данные из базы данные MongoDB и сохраняет их в виде файлов .csv

```python
import os
import dotenv
import pymongo
import pandas as pd
from sklearn.model_selection import train_test_split
pd.set_option('display.max_columns', None)

dotenv.load_dotenv('.env')
```

```python
# подключаюсь к базе данных
client = pymongo.MongoClient(os.getenv('DATA_BASE'))
db = client['moscow']
collection_1 = db['cian_1_room']
collection_2 = db['cian_2_rooms']
collection_3 = db['cian_3_rooms']
```

```python
# считываю данные из соответствующий коллекций и сохраняю их в pandas DataFrame
cian_1_room = collection_1.find()
df_1 = pd.DataFrame.from_records(cian_1_room)
df_1 = df_1.drop('_id', axis=1)

cian_2_rooms = collection_2.find()
df_2 = pd.DataFrame.from_records(cian_2_rooms)
df_2 = df_2.drop('_id', axis=1)

cian_3_rooms = collection_3.find()
df_3 = pd.DataFrame.from_records(cian_3_rooms)
df_3 = df_3.drop('_id', axis=1)
```

```python
# объединяю данные в один датафрейм
df = pd.concat([df_1, df_2, df_3])
```

```python
df.shape
```

```python
# отделяю целевую переменную
data = df.drop('price', axis=1)
target = df.price
```

```python
# разбиваю на тренировочный и тестовый датафреймы
train_data, test_data, train_target, test_target = train_test_split(data, target, test_size=0.3)
```

```python
# сохраняю в виде файлов .csv для дальнейшего использования
train_data.to_csv('./data/train_data.csv', index=False)
test_data.to_csv('./data/test_data.csv', index=False)
train_target.to_csv('./data/train_target.csv', index=False)
test_target.to_csv('./data/test_target.csv', index=False)
```
